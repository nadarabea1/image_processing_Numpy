```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy.ndimage import rotate
```

```python
myImage = cv2.imread('image.png')
img2= cv2.imread('image2.png')
```

```python
# RGB IMAGE TO GRAY SCALE
gray_img = cv2.cvtColor(myImage, cv2.COLOR_BGR2GRAY)
cv2.imshow('Image Show',myImage)
cv2.imshow("image",gray_img) 
plt.imshow(gray_img,cmap='gray') 
```

```python
# BINARY IMAGE
th, image_Binary = cv2.threshold(gray_img, 110, 255, cv2.THRESH_BINARY)
plt.figure(1); 
plt.imshow(image_Binary,cmap='gray'); 
plt.title("binary image")
```

```python
# CROPPED IMAGE
print(myImage.shape)
cropped_image = myImage[10:400, 10:550]
# cv2.imshow('Image Show',cropped_image)
plt.imshow(cropped_image,cmap='gray'); 
plt.title("cropped image")
```

```python
#RESIZE IMAGE
resize_image=cv2.resize(myImage, dsize=(200, 240), interpolation=cv2.INTER_CUBIC)
plt.imshow(resize_image,cmap='gray') 
plt.title("resize image")
```

```python
#Bluring IMAGE WITH GAUSSIAN FILTER
bluring_image = cv2.GaussianBlur(myImage, (15, 15), cv2.BORDER_DEFAULT)
plt.imshow(bluring_image,cmap='gray') 
plt.title("Blur image")
```

```python
#SHARP EDGE
# kernel=np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
# sharp_image=cv2.filter2D(src=img2,ddepth=1,kernel=kernel)
sharp_image = cv2.Canny(myImage,100,200)
# img3= cv2.resize(sharp_image, (500, 500))
# img= cv2.resize(myImage, (500, 500))
# final_img = cv2.addWeighted(img, 0.9, img3, .3, 0.0)
plt.imshow(sharp_image,cmap='gray') 
plt.title("Sharpped image")
```

```python
#SECOND IMAGE
gray_img2 =cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
plt.imshow(img2[:,:,::-1])
```

```python
#Rotate Image
Rotated_image =rotate(img2, 40)
cv2.imshow("Rotated", Rotated_image)
plt.imshow(Rotated_image,cmap='gray') 
```

```python
#Detect Edges
edge_img = cv2.Canny(img2,100,200)
cv2.imshow("Detected Edges", edge_img)
plt.imshow(edge_img,cmap='gray') 
```
